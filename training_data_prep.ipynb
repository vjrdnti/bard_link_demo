```python
import json
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
def convert_from_json_to_dataframe(file_path, record_path=['data', 'paragraphs', 'qas', 'answers']):
    file = json.loads(open(file_path).read())
    js = pd.json_normalize(file, record_path)
    m = pd.json_normalize(file, record_path[:-1])
    r = pd.json_normalize(file, record_path[:-2])
    index = np.repeat(r['context'].values, r.qas.str.len())
    m['context'] = index
    data = m[['id', 'question', 'context', 'answers']].set_index('id').reset_index()
    data['context_id'] = data['context'].factorize()[0]
    return data
    

train_file_path = './train-v2.0.json'
df = convert_from_json_to_dataframe(train_file_path)
df
```

```python
import gensim
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import torch
from transformers import AutoTokenizer, AutoModel
from transformers import GPT2Tokenizer, GPT2Model
from scipy.spatial.distance import cosine
```

```python
def calculate_lsa_score(question, context):
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform([question, context])
    svd = TruncatedSVD(n_components=2)
    lsa_matrix = svd.fit_transform(tfidf_matrix)
    lsa_score = 1 - cosine(lsa_matrix[0], lsa_matrix[1])
    return lsa_score


import gensim.downloader as api
word_vectors = api.load("glove-wiki-gigaword-100")
#word_vectors = KeyedVectors.load_word2vec_format('glove-wiki-gigaword-200', binary=False)

def calculate_embedding_score(question, context):
    question_vec = np.mean([word_vectors[word] for word in question.split() if word in word_vectors], axis=0)
    context_vec = np.mean([word_vectors[word] for word in context.split() if word in word_vectors], axis=0)
    embedding_score = 1 - cosine(question_vec, context_vec)
    return embedding_score
```

```python
dfx = df.iloc[0:10,:]
dfx.head()
```

```python
# Apply functions to each row in the DataFrame
dfx['lsa_score'] = dfx.apply(lambda x: calculate_lsa_score(x['question'], x['context']), axis=1)
dfx['embedding_score'] = dfx.apply(lambda x: calculate_embedding_score(x['question'], x['context']), axis=1)

# Print the DataFrame with added scores
dfx.head(10)
```
